```python
import numpy as np
import imageio
import matplotlib.pyplot as plt

img = imageio.imread("To Feed or Not To Feed.jpg")

img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
img_type = img.dtype

img_flip_horizontal = np.zeros(img.shape, img_type)
img_flip_vertical = np.zeros(img.shape, img_type)

for y in range(0, img_height):
    for x in range(0, img_width):
        for c in range(0, img_channel):
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c]

for y in range(0, img_height):
    for x in range(0, img_width):
        for c in range(0, img_channel):
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c]


fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

axes[0].imshow(img_flip_horizontal)
axes[0].set_title("Flip Horizontal")

axes[1].imshow(img_flip_vertical)
axes[1].set_title("Flip Vertical")

plt.show()
```
